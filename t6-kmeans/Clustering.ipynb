Basic setup:

   * Data points: $X=[x_1, x_2, \ldots, x_N]$, where $x_i\in D^n$.
   * Clusters: $\{C_1,\ldots, C_K\}$ s.t. $\cup_i C_i=\{1,\ldots, N\}$ and $C_i \cap C_j = \emptyset$.
   * Distance measure: point to piont distance $d(x_i,x_j)$; point to cluster distance $d(x_i,C_j)$. 
   * Objective: Find $K$ and $\{C_1,\ldots, C_K\}$ that minimize cummulated distance: 
   $\sum_{j \le K}\sum_{i \le C_j}d(x_i, C_j)$.

The instantiation of K-means:

   * Real-valued features: $D=\mathbb{R}$.
   * **Squared** Euclidean distnace ($l_2$-distance): $d(x_i,x_j)=||x_i-y_i||^2_2$.
   * Point to cluster distance:
      * Centroid: $c_j \in \mathbb{R}^n$ that minimizes $\sum_{i \in C_j} d(x_i,c_j)$.
      * There exists a closed form for centroid under Eucliden distance:
      $c_j=\frac{1}{|C_j|}\sum_{i \in C_j} x_i$ (the mean)

There are two major ways to define point-to-cluster distance:
   
   * Centroid: $c_j \in \mathbb{R}^n$ that minimizes $\sum_{i \in C_j} d(x_i,c_j)$.
   * Clustertroid: $c_j \in C_j$ that minimizes $\sum_{i \in C_j} d(x_i,c_j)$.
   
Note the difference is just the domain of $c_j$.

## A synthesized data set

```python
import numpy as np
import random

def gen_one_cluster(center, stdev, n):
    x_0, y_0 = center
    x = x_0 + stdev * randn(1, n)
    y = y_0 + stdev * randn(1, n)
    m = matrix(concatenate((x,y), axis=0))
    # each column is a data point
    return m
```

```python
def gen_clusters(cluster_params):
    X = ndarray((2,0)) # an empty 2-by-0 matrix
    cluster_g = [] # _g is short for ground-truth
    for (i, c) in enumerate(cluster_params):
        X_part = gen_one_cluster(**c)
        X = concatenate((X, X_part), axis=1)
        cluster_g += [i] * size(X_part, 1)
    cluster_g = array(cluster_g)
    return X, cluster_g
```

```python
cluster_params = [
            {'center': (-1,0),
             'stdev': 0.2,
             'n': 100},
            {'center': (0,0),
             'stdev': 0.2,
             'n': 100},
            {'center': (-0.5,1),
             'stdev': 0.2,
             'n': 100},
            ]

# Draw all the data points
X, cluster_g = gen_clusters(cluster_params)
scatter(array(X[0, :]), array(X[1, :]))
axis([-2, 2, -2, 2])

figure()

# Draw the ground truth with different color
# Note that cluster_g is ground-truth we get from data generator
# We will run K-means and try to find `cluster` ourselves.
# We can use the same way to draw intermediate results
scatter(array(X[0, cluster_g==0]), array(X[1, cluster_g==0]), color='b')
scatter(array(X[0, cluster_g==1]), array(X[1, cluster_g==1]), color='r')
scatter(array(X[0, cluster_g==2]), array(X[1, cluster_g==2]), color='g')

axis([-2, 2, -2, 2])
```

## K-means

This section lets you play with the K-means algorithm interactively.
Each time you want to start:

1. Execute the `# Initialization` block.
2. Execute the `# One iteration` block repeatedly. Pause and analyze at any time.

```python
def visualize_clusters(iteration, X, cluster):
    print 'iteration:', iteration
    color_band = ['b', 'r', 'g', 'k'] # add more if needed
    # plot each cluster
    for c in set(cluster):
        scatter(array(X[0, cluster==c]), array(X[1, cluster==c]), color=color_band[c])
    # plot the centroids
    for c in set(cluster):
        c_x = mean(array(X[0, cluster==c]))
        c_y = mean(array(X[1, cluster==c]))
        plot(c_x, c_y, color=color_band[c], marker='D', markersize=10)
```

```python
# Initialization
iteration = 0
n = size(X, 0)
N = size(X, 1)
# randomly assign data points to the clusters
cluster = fix(rand(N) * len(cluster_params)).astype('int32')
visualize_clusters(iteration, X, cluster)

# Note the centroids are plotted as diamond
```

```python
# One iteration
# Execute this block multiple times to see the result

iteration += 1

# Compute centroids
centroids = []
for c in set(cluster):
    c_x = mean(array(X[0, cluster==c]))
    c_y = mean(array(X[1, cluster==c]))
    centroids.append((c_x, c_y))

# Re-assign points to clusters
for i in range(0, N):
    dists = []
    # Compute distance to all centroids
    for (c, coord) in enumerate(centroids):
        d = (coord[0] - X[0,i])**2 + (coord[1] - X[1,i])**2
        dists.append((c, d))
    # assign the point to nearest centroid
    min_c, min_coord = min(dists, key=lambda x: x[1])
    cluster[i] = min_c
    
visualize_clusters(iteration, X, cluster)
```

The above is only after one iteration.
Keep on executing it. 
Usually, it converges in three iterations for this simple case.

**NOTE**:

   * See any difference of this implementation from other sources?
   In classical texts, you initialize with **centroids**.
   Here we initialize with **cluster assignments**.
   The two methods are equivalent.
   K-means iteratively compute centroids and cluster assignments.
   It converges eventually regardless of which one you start with.
   * Randomly initializing cluster assignments may not be a good idea in practice.
   K-means can be sensitive to initialization methods.
   * This implementation lends itself to more variants.
   In some distance measure, there can be no **centroids** or **clustertroids**.
   These two are just a convenient way to measure the distance between one point and a cluster.
   One can specify point to cluster distance $d(x_i,C_j)$ directly.

## Incremental K-means

The above flow is referred to as batch mode.
One drawback is that you must pass all the data points once before updating centroids.
If data points comes in one-by-one, you need to wait until the end of input stream.

K-means can also operate in incremental mode, which solves the above problem.

Suppose we alreay have reasonable centroids $c_1, \ldots, c_K$ and cluster assignment $C_1, \ldots, C_K$.
Now the $N+1$-th data point (denote as $x$ for short) comes in our system.
How to merge it with our existing clustering result?
One simple way is:

   * Assign $x$ to the nearest cluster, $C_j$. 
   * Update $c_j$, the centroid of $C_j$.
   
The first step takes only $O(K)$ time, which is efficient in most applications.
The second step needs more care.
When there are a lot of data points in the system, 
it's inefficient to compute the mean straightforwardly, which takes $O(|C_j|)$ time.
We need to incrementally update the centroid.

   * The old centroid is computed as: $c_j=\frac{1}{|C_j|}\sum_{i \in C_j} x_i$
   * The new centroid (after assigning $x$ to $C_j$) is: 
   $c'_j=\frac{1}{|C_j|+1}((\sum_{i \in C_j} x_i) + x)$
   * We can rearrange it as 
   $c'_j=\frac{1}{|C_j|+1}\sum_{i \in C_j} x_i + \frac{1}{|C_j|+1}x =  \frac{|C_j|}{|C_j|+1}\frac{1}{|C_j|}\sum_{i \in C_j} x_i + \frac{x}{|C_j|+1} = \frac{|C_j|}{|C_j|+1}c_j + \frac{x}{|C_j|+1}$
   
The new way to update centroids only takes $O(1)$!

To get more insights, let's re-arrange the centroid updating formula:

$$c'_j= \frac{|C_j|+1}{|C_j|+1}c_j - \frac{1}{|C_j|+1}c_j  + \frac{x}{|C_j|+1} = c_j + \frac{x-c_j}{|C_j|+1}$$

We see that the new data point is driving the centroid towards its own direction.
Towards this end, the second term is like a "momentum".

Now we know how to update the centroid is one new data point is assigned to a cluster.
What a set of data points $S$ are assigned to one cluster?
Actually, we have a similar result:

$$c'_j= \frac{|C_j|}{|C_j|+|S|}c_j + \frac{|S|}{|C_j|+|S|}\frac{1}{|S|}\sum_{x\in S} x$$

**EXERCISE**:
Derive the above formula

What can we learn from the formula?
It is not only a way to incrementally update centroid with arbitrary number of new data points, 
but also tells us how information can be compressed:

   * Originally, we have to record: $(c_j,|C_j|,S)$.
   The length is $1+1+|S|$
   * Now we only need to record: $(c_j,|C_j|,|S|,\frac{1}{|S|}\sum_{x\in S} x)$.
   Note that the last element is just the mean of $S$.
   The length is $4$, a constant!
   
How is this useful?
Suppose your data points come in group by group.
You don't need to record the details of each group.
Just record the the size and mean of each group.
Given this information, you will be able to update the centroids eventually.

```python
def assign_cluster(X, centroids):
    N = size(X, 1)
    cluster = [-1] * N
    for i in range(0, N):
        dists = []
        # Compute distance to all centroids
        for (c, coord) in enumerate(centroids):
            d = (coord[0] - X[0,i])**2 + (coord[1] - X[1,i])**2
            dists.append((c, d))
        # assign the point to nearest centroid
        min_c, min_coord = min(dists, key=lambda x: x[1])
        cluster[i] = min_c
    return array(cluster)

def visualize_centroids(iteration, X, centroids):
    # We visualize by centroids, not cluster assignments
    # Compare it with the previous one
    print 'iteration:', iteration
    color_band = ['b', 'r', 'g', 'k'] # add more if needed
    # plot each cluster defined by the centroids
    cluster = assign_cluster(X, centroids)
    figure()
    hold(True)
    #scatter(array(X[0,:]), array(X[1, :]))
    for c in set(cluster):
        scatter(array(X[0, cluster==c]), array(X[1, cluster==c]), color=color_band[c])
    # plot the centroids
    for (c, coord) in enumerate(centroids):
        plot(coord[0], coord[1], color=color_band[c], marker='D', markersize=10)
```

```python
# Initialization
iteration = 0
n = size(X, 0)
N = size(X, 1)
# randomly initialize two centroids
centroids = [(-0.5 + randn() * 0.2, randn() * 0.2) for i in range(len(cluster_params))]
# IMPORTANT information for incremental updates
sizes = [100] * len(cluster_params)
visualize_centroids(iteration, X, centroids)

selected_X = ndarray((2,0))

# Note the centroids are plotted as diamond
```

```python
# One iteration
# Execute this block multiple times to see the result

iteration += 1

def update_centroid_incremental(centroid, size, new_points_mean, new_points_num):
    # The general incremental update formula
    size = size / 2.0
    new_centroid = (size * matrix(centroid).T + new_points_num * matrix(new_points_mean)) / (new_points_num + size)
    return (new_points_num + size), (new_centroid[0, 0], new_centroid[1, 0])
    
# 30 random points, incremental update
x = X[:, np.random.randint(0, size(X, 1), size=30)]
# Assign it to the nearest cluster
cluster = assign_cluster(x, centroids)
for c in set(cluster):
    sizes[c], centroids[c] = update_centroid_incremental(centroids[c], sizes[c], mean(x[:, cluster==c], 1), 1)
selected_X = concatenate((selected_X, x), 1)

visualize_centroids(iteration, X, centroids)
figure()
# Plot the selected points
# You will find that as selected points become denser, the clustering result becomes more accurate.
scatter(array(selected_X[0, :]), array(selected_X[1, :]))
```

## Remark on Incremental Update

The incremental update formula make it easy to perform K-means in parallel.
Given this formula, one can readily implement K-means on MapReduce.
Furthermore, the information compression perspective makes it more powerful.
In essence, it tells you that a set of points can be **summarized** and **combined** in arbitrary order.
Search for "mergeable summary" for a recent hot topic in big data research.
With this property, we can design a multi-tier K-means:

   * In Hadoop v1, there is one component called "combiner", which gives you one more layer.
   * [Dremel](http://static.googleusercontent.com/media/research.google.com/en//pubs/archive/36632.pdf)
   allows you to summarize the data using multiple intermediate layers.
   You can think of the intermediate layers as many combiners.
   * Hadoop v2 generalizes Map-Combine-Reduce job to [DAG](http://en.wikipedia.org/wiki/Directed_acyclic_graph) job.
   * [Apache Storm](http://storm.incubator.apache.org/) is a streaming framework.
   You can apply incremental update formula to many streams to process large volume data in real time.
