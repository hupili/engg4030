## Create graph and operate on nodes/edges

````python
import networkx as nx
````

````python
# Create graph
G = nx.Graph()
````

````python
# Add single node
G.add_node(1)
G.nodes()
````

````python
# Add list of node
G.add_nodes_from([2,3])
G.nodes()
````

````python
# Nodes can be any (hashable) type
G.add_nodes_from(['u', 'v'])
G.nodes()
````

````python
# Remove node
G.remove_node('u')
G.nodes()
````

````python
# Add edges
G.add_edge(1, 2)
print G.edges()
G.add_edges_from([(2, 3), (1, 'v')])
print G.edges()
````

````python
print G.number_of_nodes()
print G.number_of_edges()
````

````python
# Get the adjacency list of one node
print G.neighbors(1)
# Get the adjacency list of all
print G.adjacency_list()
````

````python
# The adjacency_list() is not very useful without knowing the vertex
print G.nodes()
# Combine vertex label and adjacency_list
print dict(zip(G.nodes(), G.adjacency_list()))
# Equivalent information can be get by:
print nx.to_dict_of_dicts(G)
# Equivalent information can also be get by:
print G.edge
````

## Graph visualization

We play with the LegCo data again...

````python
url_legco_hk_vote_matrix = 'https://course.ie.cuhk.edu.hk/~engg4030/tutorial/tutorial7/votes-matrix.csv'
import pandas as pd
df = pd.io.parsers.read_csv(url_legco_hk_vote_matrix, index_col='member')
print 'number of users:', len(df.index)
print 'number of items:', len(df.columns)
df[:5]
````

````python
# Each column of DataFrame is a Series
# It's easier to retrieve and process series by series
dft = df.transpose()
dft[:5]
````

````python
# We can use this way to measure similarity between two members
print sum(dft['Abraham SHEK'] == dft['Alan LEONG'])
print len(dft)
````

````python
# We build graph from scratch to review the basic operations
legcohk = nx.Graph()
# Add nodes
for n in dft.columns:
    legcohk.add_node(n)
nx.draw(legcohk)
````

````python
import itertools
# itertools saves you some typing from writing nested loops
for (u, v) in itertools.combinations(legcohk.nodes_iter(), 2):
    w = float(sum(dft[u] == dft[v])) / len(dft)
    #w = float(sum(dft[u] == dft[v]))
    legcohk.add_edge(u, v, weight=w)
````

````python
#%pylab inline
#%pylab
nx.draw(legcohk)
# Now it is a complete graph.
# There are too many edges cluttering the graph.
````

````python
# In order to cut some edges, we first get an understanding of the weight distribution.
# Recall how we iterated edges before.
weights = []
for n,nbrs in legcohk.adjacency_iter():
    for nbr,eattr in nbrs.items():
        weights.append(eattr['weight'])
weights = array(weights)
hist(weights)
````

````python
#%pylab
#%pylab inline
legcohk_cut = nx.Graph()
legcohk_cut.add_nodes_from(legcohk.nodes_iter())
for n,nbrs in legcohk.adjacency_iter():
    for nbr,eattr in nbrs.items():
        if eattr['weight'] > 0.5:
            legcohk_cut.add_edge(n, nbr, eattr)
nx.draw(legcohk_cut)
# You can pop out the plot and zoom in
# The following visualization is more readable now
````

````python
pos = nx.spring_layout(legcohk_cut, dim=3, iterations=500, k=1.0/4)

figure()
hold(True)
for (k, v) in pos.iteritems():
    plot(v[0], v[1], '.b')
````

````python
#%pylab
#%pylab inline

pos = nx.spring_layout(legcohk_cut, dim=3, iterations=500, k=1.0/4)
m = matrix(pos.values()).T
x = array(m[0, :]).squeeze()
y = array(m[1, :]).squeeze()
z = array(m[2, :]).squeeze()

from mpl_toolkits.mplot3d import Axes3D
fig = figure(figsize=(15, 10))
#fig = figure()
ax = fig.add_subplot(111, projection='3d')
# Draw nodes
ax.scatter(x, y, z, s=100)
# Try to implement the picker method by referencing PCA tutorial
#ax.scatter(x, y, z, picker=True, s=100)

# Draw edges
for e in legcohk_cut.edges_iter():
    pos0 = pos[e[0]]
    pos1 = pos[e[1]]
    l = zip(pos0, pos1)
    ax.plot(l[0], l[1], l[2], color='blue', alpha=0.3)
````

## Centrality measures

[Centrality](http://en.wikipedia.org/wiki/Centrality) is a classical concept in graph analysis.
It measures the "importance" of nodes.
The notions of "importance" are different.

We only provide some samples in following sections.

**EXERCISE**: 
Refer to the [documentation](http://networkx.github.io/documentation/latest/reference/algorithms.centrality.html)
and online resources to understand those centrality measures.
Try other centrality measures that are not covered in this tutorial.
What interesting findings can you get?

````python
G = legcohk_cut
````

````python
import heapq

def top_items(G, centrality):
    nodes = heapq.nlargest(5, centrality, key=lambda n: centrality[n])
    for n in nodes:
        print n, centrality[n]
````

````python
top_items(G, nx.degree_centrality(G))
````

````python
top_items(G, nx.closeness_centrality(G))
````

````python
top_items(G, nx.betweenness_centrality(G))
````

````python
top_items(G, nx.eigenvector_centrality(G))
````

## HITS

[HITS](http://en.wikipedia.org/wiki/HITS_algorithm) works on a bi-partite graph.
We build it from the voting records.
If a member vote 'Yes' for a certain topic, we draw an edge from the member to the topic.

**EXERCISE**:
What other ways can you bulid the graph?
What is the corresponding interpretation?
Any interesting findings?

````python
voteG = nx.DiGraph()
voteG.add_nodes_from(dft.index)
voteG.add_nodes_from(dft.columns)
for member in dft.columns:
    for (topic, vote) in dft[member].iteritems():
        #print member, topic, vote
        if vote == 'Yes':
            voteG.add_edge(member, topic)

# sum of numbers and topics
print voteG.number_of_nodes()
print voteG.number_of_edges()
````

````python
hubs, authorities = nx.hits(voteG)
print "Hubs:"
top_items(G, hubs)
print "Authorities:"
top_items(G, authorities)
````
