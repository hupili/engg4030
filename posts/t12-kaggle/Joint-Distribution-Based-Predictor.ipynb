## Estimate Joint Distribution

```python
import pandas as pd
```

```python
data_train = pd.read_csv('train.csv')
data_train[:5]
```

```python
def to_categorical(data, dropna):
    newdata = pd.DataFrame(data, index=data.index)
    newdata = newdata.drop(['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], 1)
    if dropna:
        newdata = newdata.dropna()
    else:
        # For test set, we can not drop NaN values.
        # We must make predictions for everyone in order to meet the formating requirement.
        # In that case, we fill NaN by non existing categories so as to fail the table look-up step.
        newdata = newdata.fillna(-1)
    newdata['Sex'] = newdata['Sex'].map({'male': 1, 'female': 0})
    # Note that the infinite data assumption does not hold in our case.
    # We cap the Age and Fare category to avoid too many categories with inadequate data.
    # We simply demonstrate 5 categories as follows.
    # You can fine-tune categorization in other ways.
    newdata['Age'] = newdata['Age'].map(lambda x: min(int(x / 10.0), 4))
    newdata['Fare'] = newdata['Fare'].map(lambda x: min(int(x / 10.0), 4))
    return newdata

data_train_cat = to_categorical(data_train, dropna=True)
data_train_cat[:5]
```

```python
joint_frequency_total = data_train_cat.groupby(['Pclass', 'Sex', 'Age', 'Fare']).count()['PassengerId']
joint_frequency_total[:10]
```

```python
joint_frequency_survived = data_train_cat[data_train_cat['Survived'] == 1
                                          ].groupby(['Pclass', 'Sex', 'Age', 'Fare']).count()['PassengerId']
joint_frequency_survived[:10]
```

```python
import itertools
all_combinations = itertools.product(
    (1, 2, 3), # Pclass
    (0, 1), # Sex
    (0, 1, 2, 3, 4), # Age
    (0, 1, 2, 3, 4) # Fare
)

joint_distribution = pd.DataFrame(index=all_combinations)
joint_distribution['total'] = joint_frequency_total
joint_distribution['survived=1'] = joint_frequency_survived
joint_distribution['probability'] = joint_frequency_survived / joint_frequency_total
joint_distribution[:10]
```

```python
# Apply the optimal predictor (under the two assumption)
# Note, NaN is also made 0.
# This is because we have the prior knowledge that more people die than live.
# Without further information, predicting death is more reasonable.
joint_distribution['prediction'] = joint_distribution['probability'].map(lambda x: 1 if x > 0.5 else 0)
joint_distribution[:10]
```

## Prediction

After getting the joint distribution, the prediction is just like a table lookup.

```python
data_test = pd.read_csv('test.csv')
data_test_cat = to_categorical(data_test, dropna=False)
data_test_cat[:5]
```

```python
d = data_test_cat.set_index(['Pclass', 'Sex', 'Age', 'Fare'])
# This is similar to the join of tables in database.
# The two tables are indexed by the combinations of four categories.
# The join directly makes predictions.
prediction = d.join(joint_distribution)
prediction['Survived'] = prediction['prediction']
prediction[:10]
```

```python
prediction[['PassengerId', 'Survived']].to_csv('submission.csv', index=False)
```

## Exercise

You'll find that this predictor is worse than the previous sex-only predictor.
It gets a score of `0.66507`.
The main reason is that the two assumptions do not hold.
Actually, the data point is very sparse.
We have about 900 data points but $3 \times 2 \times 5 \times 5 = 150$ combinations.
Some cells do not get enough data points.
There several directions to explore:

   * Use less features. For example, if we drop the `Fare` dimension, the score goes up to `0.72727`
   * Fine-tune the categorization according to the distribution.
   For example, the interesting categorization for `Age` may not be linear.
   Instead, it could be `[0, 5]`, `[5, 15]`, `[15, 35]`, `[35, 80]` (just a guess).
   * Fill in missing values.
   Note that we simply dropped the NaN values in the pre-processing stage. 
   One data point is dropped if there is one NaN dimension.
   We can fill in the NaN with some reasonable domain-specific values, e.g. the average Age.
   * Marginalization.
   We have the statistics for all combinations of `(Pclass, Sex, Age, Fare)`.
   For a test point, we predict 0 if one of the element is missing, due to a table look-up failure.
   However, there exist better guesses based on marginalized distribution.
   For example, suppose we know a point has `Sex==female` and other dimensions are missing.
   We may better guess `Survived=1`.
