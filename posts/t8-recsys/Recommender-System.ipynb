## Recommender system

The notation:

   * Set of users: $U$
   * Set of items: $I$
   * Set of items rated by user $u$: $I(u)$.
   * Set of users who rate item $i$: $U(i)$.
   * The rating records: $R$, a set of triples $(u, i, r_{ui})$. $r_{ui}$ is the rating user $u$ give to item $i$.
   
The overall objective of a recommender system is to predict unknown ratings that user $u$ may give to item $i$.
Formally,

   * Given user $u$ and item $i$, we predict the rating by a function: $f(u,i)$.
   * Choose a loss function $L(\cdot, \cdot)$ to measure the difference between ground-truth (those in $R$) and prediction.
   * Get an optimization problem: $\text{minimize}_f L(R, f)$.

### a) The instantiation of $L$:

[SSE](http://en.wikipedia.org/wiki/Residual_sum_of_squares) (sum of squared errors) is the most commonly used metric.

$$L(R, f) = \sum_{(u, i, r_{ui}) \in R} (r_{ui} - f(u, i))^2$$

The advantage is that SSE is easy to differentiate.

### b) The instantiation of $f$

An example of a trivial predictor $f(u, i)=3$.
One can show that the [RMSE](http://en.wikipedia.org/wiki/Root-mean-square_deviation) is less than 2 if the rating scale is 1-5.

**EXERCISE**: 
Prove the above argument.

A slightly advanced predictor may be: $f(u, i) = \frac{1}{|R|} \sum_{(u, i, r_{ui}) \in R} r_{ui}$.
That is, we use the global average rating to predict unknown ones.
In practice, this may not be very bad.
Recall from Netflix data, this simple predictor can get RMSE slightly more than 1.

Now we incorporate user/item biases:

   * $f(u,i) = \mu + \alpha_u + \beta_i$
   * $\mu = \frac{1}{|R|} \sum_{(u, i, r_{ui}) \in R} r_{ui}$ is the global bias
   * $\alpha_u = \frac{1}{|U(u)|} \sum_{j \in U(u)} r_{uj}$ is the user bias
   * $\beta_i = \frac{1}{|I(i)|} \sum_{v \in I(i)} r_{vi}$ is the item bias

Further incorporate user/item "latent factors":

   * $f(u,i) = \mu + \alpha_u + \beta_i + p^T_u q_i$
   * $p_u$: the user latent factor, a $k$ dimensional vector
   * $q_i$: the item latent factor, a $k$ dimensional vector

**NOTE**:
One research direction is to build better predictor $f$. 
For example, by adding some neighbourhood information, we can get the [SVD++](http://www.recsyswiki.com/wiki/SVD%2B%2B) model.

### c) Regularization

We used a shorthand $f(\cdot, \cdot)$ for predictor in the above discussion.
The function is often characterized by a set of parameters $\theta$.
In the last instantiation of $f$, $\theta=\{\mu\}\cup\{\alpha_u, p_u: u \in U\} \cup \{ \beta_i, q_i: i \in I \}$.
The loss function minimization approach is:

$$\text{minimize}_{\theta} L(R, f_{\theta})$$

It is very easy to overfit the model.
We can add regularization term to the formulation to alleviate the problem.

$$\text{minimize}_{\theta} L(R, f_{\theta}) + \lambda P(\theta)$$

In some fields, the 2nd term is also called "penalize function".
The squared 2nd norm is often adopted for $P$.

**NOTE**:
One research direction is to incorporate side information by regularization.
For example, one user may have similar taste as his friends.
If we can get another **independent** social network (not extracted from the rating matrix),
we can add regularization terms on the edges to force user latent factors become similar.

**EXERCISE**:
What other ways can you regularize the parameters?

One possible (well working) instantiation of the whole optimization problem combining above examples of loss function and regularization is: 

$$\text{minimize}_{\theta} \sum_{(u, i, r_{ui}) \in R} (r_{ui} - (\mu + \alpha_u + \beta_i + p^T_u q_i))^2 + \lambda (\mu^2 + ||\alpha||^2 + ||\beta||^2 + \sum_{u \in U}||p_u||^2 + \sum_{i \in I}||q_i||^2)$$


### d) Training and testing set

In the above discussion, we used all the rating records, $R$, in the formulation.
It is just a shorthand for the records we feed into the optimization problem.
In practice, we need to split the records into training set, $R_{\text{train}}$, and testing set, $R_{\text{test}}$.
We feed the optimization problem by $R_{\text{train}}$ and validate the result on $R_{\text{test}}$.

### e) Super parameters

Note that the regularization coefficient $\lambda$ in the above model is not part of optimization variables $\theta$.
It can be determined by:

   * Domain knowledge.
   * Model selection techniques like [cross validation](http://en.wikipedia.org/wiki/Cross-validation_%28statistics%29).

## Data Loading

We use the legco data cralwed in [tutorial7](https://course.ie.cuhk.edu.hk/~engg4030/tutorial/tutorial7/).
We convert the voting results to numeric values in the same way.

````python
url_legco_hk_vote_matrix = 'https://course.ie.cuhk.edu.hk/~engg4030/tutorial/tutorial7/votes-matrix.csv'
import pandas as pd
df = pd.io.parsers.read_csv(url_legco_hk_vote_matrix, index_col='member')
print 'number of users:', len(df.index)
print 'number of items:', len(df.columns)
df[:5]
````

````python
# Convert votes to numeric values
def vote_to_numeric(v):
    if v == 'Yes':
        return 1
    elif v == 'No':
        return -1
    return 0
df = df.applymap(vote_to_numeric)
df[:5]
````

## Preprocessing

We need to convert the data to `crab`'s format.
First import some sample dataset to check the format.

````python
from scikits import crab
from scikits.crab import datasets
movies = datasets.load_sample_movies()
songs = datasets.load_sample_songs()
songs
````

````python
all_data_points = []
for item in df.columns:
    for user in df[item].keys():
        rating = df[item][user]
        all_data_points.append((user, item, rating))
print 'len of all_data_points:', len(all_data_points)
nonzero_data_points = filter(lambda x: x[2] != 0, all_data_points)
print 'len of nonzero_data_points:', len(nonzero_data_points)
print all_data_points[:5]
````

````python
def split_data_points(data_points, n_train):
    # data_points: [(u1, i1, r1), (u2, i2, r2), ...]
    import random
    import copy
    _data_points = copy.deepcopy(data_points)
    # shuffle is an in-place operation
    random.shuffle(_data_points)
    train_data_points = _data_points[:n_train]
    test_data_points = _data_points[n_train:]
    return train_data_points, test_data_points
    
def data_points_to_crab_model(data_points):
    from scikits.crab.models import MatrixPreferenceDataModel
    from collections import defaultdict
    data = defaultdict(dict)
    for (u, i, r) in data_points:
        data[u][i] = r
    model = MatrixPreferenceDataModel(data)
    return model
````

````python
train_data_points, test_data_points = split_data_points(nonzero_data_points, int(0.9 * len(nonzero_data_points)))
print 'len of train_data_points:', len(train_data_points)
print 'len of test_data_points:', len(test_data_points)
````

````python
model = data_points_to_crab_model(train_data_points)
````

## Helper functions for algorithm evaluation 

````python
def predict(recommender, data_points):
    return array([recommender.estimate_preference(u, i) for (u, i) in data_points])
    
def trivial_activation(x):
    if x > 0.2:
        return 1
    elif x < -0.2:
        return -1
    else:
        return 0

def identity_activation(x):
    return x
    
def evaluate(recommender, 
             test_data_points, 
             activation_function=identity_activation, 
             return_data_points=False):
    # test_data_points: [(u1, i1, r1), (u2, i2, r2), ...]
    r_predict = predict(recommender, [(u, i) for (u, i, r) in test_data_points])
    v_activation_function = vectorize(activation_function)
    r_predict_activated = v_activation_function(r_predict)
    r_real = array(zip(*test_data_points)[2])
    evaluation = {}
    if return_data_points:
        evaluation['test_data_points_with_prediction'] = \
            pd.DataFrame([(d[0], d[1], rr, rp, rpa)
                          for (d, rr, rp, rpa) in 
                          zip(test_data_points, r_real, r_predict, r_predict_activated)],
                         columns = ['user', 'item', 'r_real', 'r_predict', 'r_predict_activated'])
    evaluation['n_test_data_points'] = len(test_data_points)
    evaluation['se'] = float(sum((r_predict_activated - r_real) ** 2))
    evaluation['mse'] = evaluation['se'] / len(test_data_points)
    evaluation['rmse'] = sqrt(evaluation['mse'])
    return evaluation

def interpret_evaluation(evaluation):
    d = evaluation['test_data_points_with_prediction']
    success = d[d['r_real'] == d['r_predict_activated']]
    print 'n_test_data_points:', evaluation['n_test_data_points']
    print 'se:', evaluation['se']
    print 'mse:', evaluation['mse']
    print 'rmse:', evaluation['rmse']
    print 'accuracy:', float(len(success)) / evaluation['n_test_data_points']
    # Per user fail stat.
    # You can check who is the most unpredictable one.
    # However, due to random sampling of training and testing set, the result can be different.
    #failed['count'] = failed['user'].map(lambda x: 1)
    #fg = failed.groupby('user').aggregate({'count': lambda x: len(x)}).sort('count', ascending=False)

# Following notebook explains why we do not use the %timeit magic
# http://nbviewer.ipython.org/urls/gist.githubusercontent.com/hupili/9613294/raw/36edd53780b2fc2f024d2203c525a8e2d18e2dcf/timeit.ipynb?create=1
import time
class Timer():
    def __init__(self, event_name='Time elapse'):
        self._event_name = event_name
        self._time = time.time() 
    def __enter__(self):
        return self
    def __exit__(self, *args):
        print('%s: %s' % (self._event_name, time.time() - self._time))
````

## User-based recommendation

````python
from scikits.crab.metrics import pearson_correlation
from scikits.crab.similarities import UserSimilarity
from scikits.crab.recommenders.knn import UserBasedRecommender

with Timer('UserBasedRecommender training'):
    similarity = UserSimilarity(model, pearson_correlation)
    recommender = UserBasedRecommender(model, similarity, with_preference=True)
    
with Timer('UserBasedRecommender testing'):
    e = evaluate(recommender, test_data_points, activation_function=trivial_activation, return_data_points=True)
````

One interesting question is: how many latent factors should we use?
The more latent factors, the better accuracy we get.
However, more latent factors can result in overfiting.
In the extreme case, the model simply tries to "memorize" the input data points rather than "learn" any knowledge.

**NOTE**:
Although the model is called "SVD", the algorithm has nothing to do with classical Singular Value Decomposition.
There are many models/algorithms called "SVDxxx". 
The only common property is that they assume a latent space, namely model users and items by lower dimension $p_u$ and $q_i$.
One can show that SVD is the optimal solution if: 
1) $L$ is SSE; 
2) $f(u,i)=p^T_u q_i$ ; 
3) **rating matrix if full**, i.e. $|R|=|U|\times|I|$.
This is why this type of models are historically referred to as "SVD" models.
However, the 3rd point is always false in the context of RecSys. 
Or else, we don't need to recommend anything.

## Discussions

Key observations:

   * Neighbourhood based methods are very fast for training but slow for testing.
   * Model based methods are usually the opposite.
   (However, there exists trivial models which are fast for both training and testing)
   
**NOTE**:
Neighbourhood based methods also have their own models.
For example, the $f(\cdot, \cdot)$ you write for the User-based recommender is just the "model".

Neighbourhood based method is data-driven:

   * **Store all the data points**. (as fast as input; large memory consumption)
   * Find neighbouring data points. (very slow for straight implementation)
      * Q: How to define similar points?
   * Use knowledge from neighbouring data points for prediction. (usually fast)
      * Q: How to combine their ratings?
      
Model based method is model-driven:

   * Construct a model to describe how the data points are generated. (usually closed form)
      * ("essentially, all models are wrong, but some are useful" -- George E. P. Box)
      * Q: What is a useful (and probably reasonable) model?
   * Solve optimal parameters for the model given sample data points. (slow)
   * **Store all the parameters $\theta$ and model description $f$** as knowledge. (small memory consumption)
   * Predict ratings by function evaluation: $f_{\theta}(u,i)$. (fast)
