## Tabular data extraction

We extract tabular data from those XML files and organize them into DataFrame of [Pandas](http://pandas.pydata.org/).
As a bonus, we do some simple validation and statistics of the data.

```python
import sh
from lxml import etree

votes = []

# The following code may have problem running on Windows,
# due to the use of `sh` library and the way I build file path.
# You can try to upgrade it.
for fn in sh.ls('legco-xml').split():
    fullpath = 'legco-xml/' + fn
    date = fn[8:16]
    #print fullpath, date
    doc = etree.XML(open(fullpath).read())
    for topic in doc.xpath('//legcohk-vote/meeting/vote'):
        topic_id = '%s-%s' % (date, topic.attrib['number'])
        for member in topic.xpath('individual-votes/member'):
            member_id = member.attrib['name-en'] # Use English name as ID for sipmlicity
            vote = member.xpath('vote')[0].text
            votes.append((topic_id, member_id, vote))
```

```python
import pandas as pd
df = pd.DataFrame(votes, columns = ['topic_id', 'member_id', 'vote'])
df.to_csv('votes-all.csv')
df[:10]
```

```python
# Some checking
print len(df['topic_id'].unique())
print len(df['member_id'].unique())
print len(df) -  len(df['topic_id'].unique()) * len(df['member_id'].unique())
print df['vote'].unique()
```

```python
# The difference is number of topics
# Check the total number of votes for each member
df.groupby('member_id').aggregate({'vote': lambda x: x.count()}).sort(columns=['vote'], ascending=True)[:10]
# We catch it!
# Using name as ID is not ideal becuase people change names sometimes.
```

```python
def normalize_name(t):
    (topic_id, member_id, vote) = t
    # According to the numbers, they seem to be the same person
    if member_id == 'Dr Joseph LEE':
        member_id = 'Prof Joseph LEE'
    # Other normalization if any
    return (topic_id, member_id, vote)
votes = map(normalize_name, votes)
```

```python
# Re generate pandas data frame.
df = pd.DataFrame(votes, columns = ['topic_id', 'member_id', 'vote'])
df.to_csv('votes-all.csv')
# Verify
print len(df) -  len(df['topic_id'].unique()) * len(df['member_id'].unique())
```

```python
# Let's see who vote many "Yes"
yes_count = df[df['vote']=='Yes'].groupby('member_id').aggregate({'vote': lambda x: x.count()}).sort('vote', ascending=False)
print "The Yes persons"
print yes_count[:5]
print "The No persons"
print yes_count[-5:]
hist(yes_count['vote'].values)
```

```python
# Let's see who vote many "No"

no_count = df[df['vote']=='No'].groupby('member_id').aggregate({'vote': lambda x: x.count()}).sort('vote', ascending=False)
print "The No persons"
print no_count[:15]
print "The Yes persons"
print no_count[-5:]
hist(no_count['vote'].values)
```

## Matrix form

Now we only the data in form a tuples: (topic_id, member_id, vote).
We want to transform it to a matrix for further processing.
We treat members as data points (rows) and topics as features (columns).

```python
from collections import defaultdict
topic_records = defaultdict(dict)
for (topic_id, member_id, vote) in votes:
    topic_records[topic_id][member_id] = vote
pd.DataFrame(topic_records).to_csv('votes-matrix.csv', index_label='member')
```
