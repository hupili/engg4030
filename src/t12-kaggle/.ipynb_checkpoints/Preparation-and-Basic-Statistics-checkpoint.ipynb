## Correlation

The `Age` and `Pclass` columns are categorical data.
It's good to visualize them using pie charts.


```python
figure(figsize=(12, 6))

subplot(1, 2, 1)
survived_male = data[data['Sex'] == 'male']['Survived'].value_counts().sort_index()
pie(survived_male.values)
legend(survived_male.index)

subplot(1, 2, 2)
survived_male = data[data['Sex'] == 'female']['Survived'].value_counts().sort_index()
pie(survived_male.values)
legend(survived_male.index)
```

```python
figure(figsize=(18, 6))

for i in range(1, 4):
    subplot(1, 3, i)
    survived_male = data[data['Pclass'] == i]['Survived'].value_counts().sort_index()
    pie(survived_male.values)
    legend(survived_male.index)
    title('Pclass=%d' % i)
```

`Age` and `Fare` are real-valued data.
Normally, we can use scatter plots to explore the relationships.
Since `Survived` itself is categorical, the scatter plot looks very cluttered and does not give too much information.
Instead, we plot the KDE of different categories to find any trend.

```python
data.plot(kind='scatter', x='Age', y='Survived', alpha=0.2)
```

```python
figure(figsize=(12,4))
data['Age'].dropna().plot(kind='kde')
data[data['Survived']==1]['Age'].dropna().plot(kind='kde')
data[data['Survived']==0]['Age'].dropna().plot(kind='kde')
legend(['Whole', 'Live', 'Dead'])
axis([0, 80, axis()[2], axis()[3]])
```

```python
figure(figsize=(12,4))
data['Fare'].dropna().plot(kind='kde')
data[data['Survived']==1]['Fare'].dropna().plot(kind='kde')
data[data['Survived']==0]['Fare'].dropna().plot(kind='kde')
legend(['Whole', 'Live', 'Dead'])
axis([0, 200, axis()[2], axis()[3]])
```

## Simple Predictor Based-on Only Sex

It's apparent that females has much better survival rate than males.
Here's a simple rule:

   * Predict 1 for a test point if `Sex==female`
   * Predict 0 for a test point if `Sex==male`

```python
test_data = pd.read_csv('test.csv')
test_data[:5]
```

```python
test_data['Survived'] = test_data['Sex'].apply(lambda x: {'male': 0, 'female': 1}[x])
test_data[:5]
```

```python
# Now it's time to read the submission format.
test_data[['PassengerId', 'Survived']].to_csv('submission.csv', index=False)
```
